```python
import numpy as np

gbr=cv2.imread('/content/drives/MyDrive/FILE TUGAS KAMPUS/daunjambu2.jpeg')
# konversi gmbar bgr openCV ke rgb matplotlib
gbr_bgr=cv2.cvtColor(gbr,cv2.COLOR_BGR2RGB)
# tampilkan gbr_bgr
plt.imshow(gbr_bgr)
plt.show()

# buat mask
mask=np.zeros(gbr.shape[:2], np.uint8)
mask[30:900, 220:550] = 255
# tampilkan mask
plt.imshow(mask,'gray')
plt.show()

# gabungkan gambar dan mask
mask_gbr=cv2.bitwise_and(gbr_bgr, gbr_bgr, mask = mask)
# tampilkan mask_gbr
plt.imshow(mask_gbr)
plt.show()

# lakukan perhitungan histogram untuk gambar dan masuk
hist_bgr=cv2.calcHist([gbr_bgr],[0],None,[256],[0,256])
hist_mask=cv2.calcHist([gbr_bgr],[0], mask, [256],[0,256])

# Tampilkan grafik histogram gambar dan masuk
plt.plot(hist_bgr)
plt.plot(hist_mask)
plt.show
```

Histogram yang ditampilkan di atas merupakan representasi grafis dari distribusi intensitas piksel pada citra digital. Sumbu horizontal grafik ini merepresentasikan nilai intensitas piksel, berkisar dari 0 (gelap total) hingga 255 (terang total), sedangkan sumbu vertikal menunjukkan frekuensi kemunculan atau jumlah piksel pada setiap tingkat intensitas tersebut. Terlihat dua kurva pada plot: kurva biru yang mewakili histogram citra asli secara keseluruhan (khususnya pada saluran Biru), dan kurva oranye yang menunjukkan histogram untuk area spesifik pada citra asli yang ditentukan oleh sebuah mask (juga pada saluran Biru). Analisis visual pada kurva-kurva ini mengungkapkan bahwa distribusi intensitas cenderung terkonsentrasi pada rentang menengah, khususnya di sekitar nilai 100 hingga 150, dengan puncak yang menonjol di sekitar 120. Pola ini mengindikasikan bahwa citra didominasi oleh area dengan tingkat kecerahan sedang. Kurva biru, yang mencakup seluruh gambar, menunjukkan frekuensi piksel yang lebih tinggi pada puncaknya dibandingkan kurva oranye yang hanya menganalisis area masked. Meskipun kurva oranye lebih rendah, sebarannya mungkin berbeda, memberikan wawasan tentang variasi intensitas di dalam area yang di-mask. Secara umum, histogram ini membantu kita memahami sebaran kecerahan dalam citra dan membandingkan distribusi tersebut antara seluruh gambar
